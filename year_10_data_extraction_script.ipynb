```python
import pyautogui
import keyboard
import time

def check_exit():
    if keyboard.is_pressed('x'):
        print("Exit key pressed. Stopping...")
        return True
    return False

k = 217
start_k = 217 # Remember the starting value

while k <= 226:
    if check_exit(): break
    pyautogui.click(1703, 341)         # default
    time.sleep(0.5)
    if check_exit(): break
    pyautogui.click(1737, 387)         # new shortlist
    time.sleep(0.5)
    if check_exit(): break
    pyautogui.write(f"shortlist {k}")  # write number
    time.sleep(0.5)
    if check_exit(): break
    pyautogui.press("enter")
    time.sleep(0.5)
    if check_exit(): break
    pyautogui.click(1842, 340)         # settings
    time.sleep(0.5)
    if check_exit(): break
    pyautogui.click(1781, 509)         # import
    time.sleep(0.5)
    pyautogui.click(1298, 427)         # select first position 
    time.sleep(0.5)
    
    # Press down (k - start_k) times (0 for k=10, 1 for k=11, etc.)
    for i in range(k - start_k):
        if check_exit(): break
        pyautogui.press("down")
    
    if check_exit(): break
    time.sleep(0.5)
    pyautogui.click(1420, 1002)        # click load
    time.sleep(0.5)
    if check_exit(): break
    pyautogui.press("enter")           # load the selected shortlist
    time.sleep(0.5)
    if check_exit(): break
    pyautogui.hotkey("ctrl", "p")      # print the data
    time.sleep(0.5)
    if check_exit(): break
    pyautogui.click(1069, 710)         # ok button
    time.sleep(0.5)
    if check_exit(): break
    pyautogui.doubleClick(630, 555)    # click the data folder    
    time.sleep(0.5)
    if check_exit(): break
    pyautogui.doubleClick(630, 555)    # save in the first folder
    time.sleep(0.5)
    if check_exit(): break
    pyautogui.doubleClick(939, 328)    # double click to name the data
    time.sleep(0.5)
    if check_exit(): break
    pyautogui.write(f"shortlist {k}")  # change the name 
    time.sleep(0.5)
    if check_exit(): break
    pyautogui.press("enter")           # press enter
    time.sleep(0.5)
    k += 1
```
