```python
import time
import pyautogui
import keyboard

k = 1

while k<16:
    pyautogui.click(1335, 1145)                   # shortlist button
    time.sleep(2)
    pyautogui.doubleClick(935, 603)               # the write Bar to write the shortlsit name
    time.sleep(2)
    pyautogui.write(f"shortlist {k}")             # name the shortlist  
    time.sleep(2)
    pyautogui.press("enter")                      # press enter
    time.sleep(2)
    k += 1
```

```python
import time
import pyautogui
import keyboard

k = 161
first_iteration = True

while k < 170:
    
    if keyboard.is_pressed('tab'):
        break
        
    time.sleep(1)
    pyautogui.click(1678, 346)                 # click shortlist
    time.sleep(1)
    
    if first_iteration:
        time.sleep(1)
        pyautogui.press("down")                # move down (only first time)
        first_iteration = False
    else:
        time.sleep(1)
        pyautogui.press("down")                # move down
        time.sleep(1)
        pyautogui.press("down")                # move down twice for subsequent iterations
    
    time.sleep(1)
    pyautogui.press("enter")                   # select the shortlist
    time.sleep(1)
    pyautogui.click(1828, 343)                 # settings button
    time.sleep(1)
    pyautogui.click(1744, 385)                 # save
    time.sleep(1)
    pyautogui.click(530, 771)                  # click bar
    time.sleep(1)
    pyautogui.write(f" {k}")                   # name the shortlist  
    time.sleep(1)
    pyautogui.click(1415, 1007)                # click ok
    time.sleep(1)
    pyautogui.click(1678, 346) 
    k += 1
```
